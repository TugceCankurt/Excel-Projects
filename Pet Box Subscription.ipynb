- There were only unique product IDs, as expected.
- There were 11 product categories, as expected.
- The number format in the sales column was converted as requested. 
- The format in Price column was changed from General to Number.
- Vendor ID is as expected.
- Pet size is as expected.
- I removed rows that contain "hamster" and "rabbit" as pet types which are 46 rows as requested.
- Consumer's ratings are on 10 point scale as expected.
